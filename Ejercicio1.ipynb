```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Datos de los puntos
xs = np.array([0, 1, 2])
ys = np.array([1, 5, 3])

# Crear el spline cúbico
spline = CubicSpline(xs, ys)

# Graficar los puntos y el spline
x_fine = np.linspace(0, 2, 100)
y_fine = spline(x_fine)

plt.plot(xs, ys, 'ro', label='Puntos dados')
plt.plot(x_fine, y_fine, label='Spline cúbico')
plt.title("Spline cúbico para los puntos dados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
