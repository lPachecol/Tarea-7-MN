```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Datos de los puntos
xs = np.array([-1, 1])
ys = np.array([1, 3])

# Condiciones de frontera
derivadas = [1, 2]  # f'(x0) = 1, f'(xn) = 2

# Crear el spline cúbico con las condiciones de frontera
spline = CubicSpline(xs, ys, bc_type=((1, derivadas[0]), (1, derivadas[1])))

# Graficar los puntos y el spline
x_fine = np.linspace(-1, 1, 100)
y_fine = spline(x_fine)

plt.plot(xs, ys, 'ro', label='Puntos dados')
plt.plot(x_fine, y_fine, label='Spline cúbico')
plt.title("Spline cúbico con condiciones de frontera")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
